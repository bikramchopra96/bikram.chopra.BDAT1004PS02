```python
import os
#modified the code to provide error message in case file not found.
def fileLength(fname):
    #try except for exception handling
    try:
        statinfo = os.stat(fname)
        return statinfo.st_size
    except FileNotFoundError:
        print('File ' + fname + ' not Found')
    except:
        print("File is not readable")
```

Now write a class named Kangaroo as a subclass of Marsupial that inherits all the
attributes of Marsupial and also:
a. extends the Marsupial __init__ constructor to take, as input, the
coordinates x and y of the Kangaroo object,
b. supports method jump that takes number values dx and dy as input and
moves the kangaroo by dx units along the x-axis and by dy units along the yaxis, and
c. overloads the __str__ operator so it behaves as shown below.

```python
class Marsupial:
    #creating class and making constructor
    def __init__(self):
        self.data = []

    def put_in_pouch(self,thing):
        self.data.append(thing)

    def pouch_contents(self):
        print(self.data)
        return self.data
#using the class objects of Marsupial class in the Kangaroo class
class Kangaroo(Marsupial):
    def __init__(self,x,y):
        self.x = x
        self.y = y
        Marsupial.__init__(self)

    def __str__(self):
        return "I am a Kangaroo located at coordinates ({},{})".format(self.x, self.y)

    def jump(self, dx,dy):
        self.x += dx
        self.y += dy

print('First Case')
# created m object of Marsupial
m = Marsupial()
m.put_in_pouch('doll')
m.put_in_pouch('firetruck')
m.put_in_pouch('kitten')
m.pouch_contents()

print('\nSecond Case')
# created k object of Kangaroo
k = Kangaroo(0,0)
print(k)
k.put_in_pouch('doll')
k.put_in_pouch('firetruck')
k.put_in_pouch('kitten')
k.pouch_contents()
k.jump(1,0)
k.jump(1,0)
k.jump(1,0)
print(k)
```

Question 5
Write function collatz() that takes a positive integer x as input and prints the
Collatz sequence starting at x. A Collatz sequence is obtained by repeatedly applying
this rule to the previous number x in the sequence:
x = {
洧논/2 洧녰洧녭 洧논 洧녰洧 洧뉧롐洧뉧롐뜦n3洧논 + 1 洧녰洧녭 洧논 洧녰洧 洧녶洧녬d}
Your function should stop when the sequence gets to number 1. Your
implementation must be recursive, without any loops.


```python
def collatz(x):
    if(x == 1):
        #stop if x is 1
        print(x)
        return
    elif (x % 2 == 0):
        #if the x is even
        print(x)
        #calling the same function for recursion
        collatz(x/2)
    else:
        print(x)
        #in case x is odd
        collatz(3*x + 1 )

collatz(10)
```

Question 6
Write a recursive method binary() that takes a non-negative
integer n and prints the binary representation of integer n.

```python
def binary(x):
    if( x == 0 or x == 1):
        # if 0 or 1 then same value
        print(x,end = "")
        return
    else:
        #calculating the binary values by finding remainders of every 2nd bit
        q = int(x/2)
        r = x%2
        binary(q)
        print(r,end="")

binary(0)
print("")
binary(1)
print("")
binary(3)
print("")
binary(9)
```

Question 8
Implement recursive function webdir() that takes as input: a URL (as a string) and
non-negative integers depth and indent. Your function should visit every web
page reachable from the starting URL web page in depth clicks or less, and print
each web page's URL. As shown below, indentation, specified by indent, should
be used to indicate the depth of a URL.

```python
from urllib.request import Request, urlopen
from urllib.parse import urljoin
from html.parser import HTMLParser

class Collector(HTMLParser):
    def __init__(self, url):
        HTMLParser.__init__(self)
        self.url = url
        self.links = []


    def handle_starttag(self, tag, attrs):
        if tag == 'a':
            for attr in attrs:
                if attr[0] == 'href':
                    absolute = urljoin(self.url, attr[1])
                    if absolute[:4] == 'http':  # collect HTTP URLs
                        self.links.append(absolute)

    def getLinks(self):
        return self.links



def analyze(url):
    req = Request(url, headers={'User-Agent': 'XYZ/3.0'})
    webpage = urlopen(req, timeout=10).read().decode()
    collector = Collector(url)
    collector.feed(webpage)
    urls = collector.getLinks()  # get list of links
    return urls


def webdir(url, depth,indent):
    if(depth == -1):
        return
    try:
        pages = analyze(url)    #Getting a list of webpages's url reachable from url
    except:
        return   #If the link is invalid

    for x in range(0,indent):
        print('\t',end="")
    print(url)
    for links in pages:   #Visiting webpages reachable from url
        webdir(links, depth - 1, indent + 1)


url = 'https://www.oreilly.com/library/view/web-scraping-with/9781491910283/ch13.html'
webdir(url, 2, 0)
```

Write list comprehension expressions that use list words and generate the following

a) ['THE', 'QUICK', 'BROWN', 'FOX', 'JUMPS', 'OVER', 'THE',

d) [['THE', 'the', 3], ['QUICK', 'quick', 5], ['BROWN',

length of the word.)
e) ['The', 'quick', 'brown', 'fox', 'jumps', 'over', 'the',

```python
words = ['The', 'quick', 'brown', 'fox', 'jumps', 'over', 'the', 'lazy', 'dog']
#Converting all to Capslock . Answer of (A)
caps = [w.upper() for w in words]
print(caps)

#Converting all to lower case . Answer of (B)
lowercase = [w.lower() for w in words]
print(lowercase)

#finding the lengths of each member of list. Answer of (C)
length = [len(w) for w in words]
print(length)

#Answer of (D)
combo = [[w.upper(), w.lower() , len(w)] for w in words]
print(combo)

#Filtering out with 4 or more than 4 characters. Answer of (E)
fourormore = [w for w in words if len(w) > 3]
print(fourormore)
```
